Before outlier removal:

Min: 0.123333

Std Dev: 23.320465

========================================================

The max value (2636.443333) is extremely large compared to Q3 (29.82667). This suggests potential outliers in higher values. To define a range:

Use an IQR-based rule:

Values above
Q3 + 1.5 × IQR

might be considered outliers:
𝐼𝑄𝑅 = 𝑄3 − 𝑄1 = 29.82667 − 4.303333 = 25.523334

𝑈𝑝𝑝𝑒𝑟 𝑏𝑜𝑢𝑛𝑑 = 𝑄3 + 1.5 × 𝐼𝑄𝑅 = 29.826667 + (1.5 × 25.523334) = 67.111668

𝐿𝑜𝑤𝑒𝑟_𝑏𝑜𝑢𝑛𝑑 = 𝑄1 − 1.5 × 𝐼𝑄𝑅 = 4.303333 − (1.5 × 25.523334)= −33.481668

Therefore, keep light_intensity values in the range of [0, 67.11].

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate IQR, lower, and upper bounds
Q1 = 4.303333
Q3 = 29.826667
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = max(0, Q1 - 1.5 * IQR)  # Adjusting for non-negative values

# Filter the data to remove outliers
filtered_df = mergin_cleaned_agg_df[
    (mergin_cleaned_agg_df['light_intensity'] >= lower_bound) & 
    (mergin_cleaned_agg_df['light_intensity'] <= upper_bound)
]

# Plot the boxplot for light intensity
plt.figure(figsize=(10, 6))
sns.boxplot(data=mergin_cleaned_agg_df, x='light_intensity', color='skyblue')
plt.axvline(upper_bound, color='red', linestyle='--', label=f'Upper Bound ({upper_bound:.2f})')
plt.axvline(lower_bound, color='green', linestyle='--', label=f'Lower Bound ({lower_bound:.2f})')

plt.title('Boxplot of Light Intensity with Outliers')
plt.xlabel('Light Intensity')
plt.legend()
plt.show()

# Print the new boundaries
print(f"Lower Bound: {lower_bound:.2f}, Upper Bound: {upper_bound:.2f}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate IQR, lower, and upper bounds
Q1 = 4.303333
Q3 = 29.826667
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = max(0, Q1 - 1.5 * IQR)  # Adjusting for non-negative values

# Plot the boxplot for light intensity
plt.figure(figsize=(12, 8))
sns.boxplot(data=mergin_cleaned_agg_df, x='light_intensity', color='skyblue', showfliers=False)
sns.stripplot(data=mergin_cleaned_agg_df, x='light_intensity', color='black', alpha=0.6, jitter=0.1)

# Add vertical lines for bounds
plt.axvline(upper_bound, color='red', linestyle='--', linewidth=1.5, label=f'Upper Bound ({upper_bound:.2f})')
plt.axvline(lower_bound, color='green', linestyle='--', linewidth=1.5, label=f'Lower Bound ({lower_bound:.2f})')

# Annotate Q1, Q3, and median
plt.text(Q1, -0.5, f"Q1 ({Q1:.2f})", color='blue', fontsize=10, ha='center')
plt.text(Q3, -0.5, f"Q3 ({Q3:.2f})", color='blue', fontsize=10, ha='center')
median = mergin_cleaned_agg_df['light_intensity'].median()
plt.text(median, -0.5, f"Median ({median:.2f})", color='purple', fontsize=10, ha='center')

# Limit x-axis range for better focus
plt.xlim(0, upper_bound * 2)

# Add titles and labels
plt.title('Boxplot of Light Intensity with Outliers', fontsize=14)
plt.xlabel('Light Intensity', fontsize=12)
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

# Print the new boundaries
print(f"Lower Bound: {lower_bound:.2f}, Upper Bound: {upper_bound:.2f}")
```

Define a Range Using the IQR Method
To detect and exclude outliers, calculate the IQR and use it to define lower and upper bounds:

Step 1: Calculate the IQR

IQR = 𝑄3 − 𝑄1 = 60140 − 12509 = 47,631

Step 2: Define Upper and Lower Bounds

Upper Bound = 𝑄3 + 1.5 × IQR = 60140 + (1.5 × 47,631) = 131,586.5

Lower Bound = 𝑄1 − 1.5 × IQR = 12509 − (1.5 × 47,631) = −58,937.5

Since internet speed cannot be negative, adjust the lower bound to 1 (the minimum possible speed in kbps).

Suggested Range

Lower Bound: 1 kbps

Upper Bound: 131,586.5 kbps
